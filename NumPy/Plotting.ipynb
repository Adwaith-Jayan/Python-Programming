```python
import numpy as np
import matplotlib.pyplot as plt
#or use from matplotlib import pyplot plt
```

```python
#Scatter plot
x=np.random.random(50)*100
y=np.random.random(50)*100

plt.scatter(x,y,color="black",s=150,marker="*")
plt.show()
```

```python
#line chart

x=[x for x in range(20)]
y=np.random.randint(1,20,size=(20))
plt.plot(x,y)
plt.show()
plt.plot(x,y,color="red",linestyle="--")
plt.show()
plt.plot(x,y,"g--")
plt.show()
```

```python
#Bar Graph
x=["C","Python","C++","Java","ASM","JS"]
y=[20,30,40,10,5,10]

plt.bar(x,y)
plt.show()

plt.bar(x,y,color="black",width=0.5)
plt.show()
```

```python
#Histogram

x=np.random.randint(1,100,size=(50))
plt.hist(x)
plt.show()
```

```python
#pie plots
x=["GO","Python","C","C++","JAVA"]
y=[5,5,10,20,22]

plt.pie(y,labels=x)
plt.show()
plt.pie(y,labels=x,explode=[1,0,0,0,0])
plt.show()
plt.pie(y,labels=x,autopct="%0.f%%")
plt.show()
```

```python
#sin wave with length, title and axis name
import math
x=np.arange(0,2* math.pi,0.05)
y=np.sin(x)

plt.plot(x,y,label="sin")
plt.title("sin Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend(loc ="upper right")
plt.show()
```

```python
#sin, cos wave 
y1=np.cos(x)
plt.plot(x,y,"r--",label="sin")
plt.plot(x,y1,color="black",linestyle="-.",label="cos")
plt.xlabel("x")
plt.ylabel("sin(x)/cos(x)/tan(x)")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```

```python
#subplots
x=np.arange(0,2*math.pi,0.05)
fig,axs=plt.subplots(1,2)
axs[0].plot(x,np.sin(x),label="sin",color="black",linestyle="--")
axs[1].plot(x,np.cos(x),label="cos")
axs[0].legend(loc="upper right")
axs[1].legend(loc="upper right")
plt.show()
```
